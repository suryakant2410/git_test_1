```python
'''
.       - Any Character Except New Line
\d      - Digit (0-9)
\D      - Not a Digit (0-9)
\w      - Word Character (a-z, A-Z, 0-9, _)
\W      - Not a Word Character
\s      - Whitespace (space, tab, newline)
\S      - Not Whitespace (space, tab, newline)

\b      - Word Boundary
\B      - Not a Word Boundary
^       - Beginning of a String
$       - End of a String

[]      - Matches Characters in brackets
[^ ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)
```

```python
text_to_search = '''\t
abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ\s
321-555-4321
1234567890
Ha\tHaHa
MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )
khanafsaan11.com
321-555-4321
123.555.1234
123*555*-1234
123.555.1234
800-555-1234
900-555-1234
    Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T
Mr_hello
'''
```

```python
import re
```

```python
# what i want to search
pattern = re.compile(r'afsaan')

# from where you want to search
matches = pattern.finditer(text_to_search)

list(matches)
```

```python
data = 'cat catherine catholic wildcat copycat uncatchable'

# animal cat
pattern = re.compile(r'cat\b')

matches = pattern.finditer(data)

list(matches)
```

```python
| - end of a word

right side of the cursor should not be a word character
left side of the cursor should be a word character
```

```python
data = 'cat catherine catholic wildcat copycat uncatchable'

# animal cat
pattern = re.compile(r'\bcat')

matches = pattern.finditer(data)

list(matches)
```

```python
| - start of a word

right side of the cursor should be a word character
left side of the cursor should not be a word character
```

```python
data = ' catherine catholic cat wildcat  copycat uncatchable '

# animal cat
pattern = re.compile(r'\bcat\b')

matches = pattern.finditer(data)

list(matches)
```

```python
\B - not a word boundry

| -- right

both the sides should be word char

| -- left

both the sides should be word char
```

```python
data = 'catherine catholic cat wildcat  copycat uncatchable '

# animal cat
pattern = re.compile(r'cat\B')

matches = pattern.finditer(data)

list(matches)
```

```python
data = 'catherine catholic cat wildcat  copycat uncatchable '

# animal cat
pattern = re.compile(r'\Bcat\B')

matches = pattern.finditer(data)

list(matches)
```

```python
# ^ - beginning of the string\

data = """is an Argentine professional footballer who plays as a forward for Ligue 1 
club Paris Saint-Germain and captains the Argentina national team. Widely 
regarded as one of the greatest players of all time, Messi has won a record seven 
Ballon d'Or awards
"""

pattern = re.compile(r'^Is')

matches = pattern.finditer(data)

list(matches)
```

```python
# $ - beginning of the string\

data = """is an Argentine professional footballer who plays as a forward for Ligue 1 
club Paris Saint-Germain and captains the Argentina national team. Widely 
regarded as one of the greatest players of all time, Messi has won a record seven 
Ballon d'Or awards
"""

pattern = re.compile(r'wardS$')

matches = pattern.finditer(data)

list(matches)
```
