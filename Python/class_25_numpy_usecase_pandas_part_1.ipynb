```python
In pandas we have two Data Structures

1.  Series ---> 1D
2.  Dataframe ---> 2D
```

Series

```python
1. list/tuple
2. array
2. dict
```

```python
# list
fifa_knockout_teams = ['argentina','france','morocoo','crotia','spain']
pd.Series(fifa_knockout_teams)
```

```python
fifa_knockout_teams = ['argentina','france','morocoo','crotia','spain']
pd.Series(fifa_knockout_teams, index=['teams1','teams2','teams3','teams4','teams5'])
```

```python
fifa_knockout_teams = ['argentina','france','morocoo','crotia','spain']
pd.Series(fifa_knockout_teams, index=['teams1','teams2','teams3','teams4'])
```

```python
# array
fifa_knockout_teams = ['argentina','france','morocoo','crotia','spain']
array_data = np.array(fifa_knockout_teams)
pd.Series(array_data)
```

```python
fifa_teams = {'teams1':'argentina','teams2':'spain','teams3':'france'}
pd.Series(fifa_teams)
```

```python
fifa_teams = {'teams1':'argentina','teams2':'spain','teams3':'france'}
pd.Series(fifa_teams, index=[1,2,'apple'])
```

```python
fifa_teams = {'teams1':'argentina','teams2':'spain','teams3':'france'}
pd.Series(fifa_teams, index=[1,2,'teams3'])
```

```python
fifa_teams = {1,2,3}
pd.Series(fifa_teams)
```

```python
# indexing

can be done in 2 ways

1. by giving the zero index value
2. by giving the custom indexing
```

```python
fifa_knockout_teams = ['argentina','france','morocoo','crotia','spain']
data = pd.Series(fifa_knockout_teams, index=['teams1','teams2','teams3','teams4','teams5'])
data
```

```python
# zero indexing

data[2]
```

```python
# custom indexing
data['teams3']
```

# slicing

```python
data[::2]
```

```python
data[1:3]

# start - inclusive
# stop - exclusive
```
